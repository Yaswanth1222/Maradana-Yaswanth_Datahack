```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

```python
y_train = pd.read_csv('Downloads/hackathon_datasets/training_set_labels.csv')
X_train = pd.read_csv('Downloads/hackathon_datasets/training_set_features.csv')
X_test = pd.read_csv('Downloads/hackathon_datasets/test_set_features.csv')
y_test = pd.read_csv('Downloads/hackathon_datasets/submission_format.csv')

X_train_copy = X_train.copy()
y_train_copy = y_train.copy()
X_test_copy = X_test.copy()
y_test_copy = y_test.copy()
```

```python
X_train.drop(columns=['respondent_id'], inplace = True)
X_test.drop(columns=['respondent_id'], inplace = True)

y_1 = y_train[['xyz_vaccine']]
y_2 = y_train[['seasonal_vaccine']]

y_test_1 = y_train[['xyz_vaccine']]
y_test_2 = y_train[['seasonal_vaccine']]
```

```python
imputer = SimpleImputer(strategy='most_frequent')
X_train_new = imputer.fit_transform(X_train)
X_test_new = imputer.transform(X_test)
X_train_new = pd.DataFrame(X_train_new, columns=X_train.columns)
X_test_new = pd.DataFrame(X_test_new, columns=X_test.columns)
```

```python
X_train_new.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
obj_col = X_train_new.select_dtypes('object').columns
col = X_train_new.columns
for column in obj_col:
    le = LabelEncoder()
    X_train_new[column] = le.fit_transform(X_train_new[column])
    X_test_new[column] = le.fit_transform(X_test_new[column])
```

Executing different models below:

Logistic Regression:

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

model_xyz = LogisticRegression(max_iter=10000)
model_xyz.fit(X_train_new, y_1)

model_seasonal = LogisticRegression(max_iter=10000)
model_seasonal.fit(X_train_new, y_2)

xyz_res = model_xyz.predict_proba(X_test_new)[:, 1] 
seasonal_res = model_seasonal.predict_proba(X_test_new)[:, 1] 

s1 = roc_auc_score(y_1, model_xyz.predict_proba(X_train_new)[:, 1],average='macro')
s2 = roc_auc_score(y_2, model_seasonal.predict_proba(X_train_new)[:, 1],average='macro')

print("ROC AUC for xyz_vaccine = ",s1)
print("ROC AUC for seasonal_vaccine = ",s2)
print("ROC AUC in Logistic = ",(s1+s2)/2)
```

Gaussian NB:

```python
from sklearn.naive_bayes import GaussianNB

model_xyz = GaussianNB()
model_xyz.fit(X_train_new, y_1)

model_seasonal = GaussianNB()
model_seasonal.fit(X_train_new, y_2)

xyz_1 = model_xyz.predict_proba(X_test_new)[:, 1] 
seasonal_2 = model_seasonal.predict_proba(X_test_new)[:, 1] 
```

```python
s1 = roc_auc_score(y_1, model_xyz.predict_proba(X_train_new)[:, 1],average='macro')
s2 = roc_auc_score(y_2, model_seasonal.predict_proba(X_train_new)[:, 1],average='macro')

print("ROC AUC for xyz_vaccine = ",s1)
print("ROC AUC for seasonal_vaccine = ",s2)
print("ROC AUC in Logistic = ",(s1+s2)/2)
```

Multinomial NB:

```python
from sklearn.naive_bayes import MultinomialNB

model_xyz = MultinomialNB()
model_xyz.fit(X_train_new, y_1)

model_seasonal = MultinomialNB()
model_seasonal.fit(X_train_new, y_2)

xyz_1 = model_xyz.predict_proba(X_test_new)[:, 1] 
seasonal_2 = model_seasonal.predict_proba(X_test_new)[:, 1] 
```
